```python
import numpy as np
import matplotlib as plt
import pandas as pd
import seaborn as sn
```

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/main/titanic-training-data.csv")
df.dtypes

```

```python
df.shape
```

```python
#WHEN TO USE relplot ??
#Relplot can be used to make realtion between numerical attributes or relation bw catagorical and number attributes.  

sn.relplot(x="PassengerId",y="Survived",data=df)
sn.relplot(x="PassengerId",y="SibSp",data=df,color="k")
```

```python
sn.relplot(x="PassengerId",y="SibSp",data=df,color="k",hue="Pclass")
sn.relplot(x="PassengerId",y="SibSp",data=df,color="k",hue="Sex")
```

```python
sn.catplot(x="Pclass",y="Survived",data=df)
sn.catplot(x="Pclass",y="SibSp",data=df)
```

```python
sn.catplot(x="Pclass",y="Age",data=df)
sn.catplot(x="Pclass",y="Age",data=df,jitter=False)
sn.catplot(x="Pclass",y="Age",data=df,jitter=False,hue="Sex")
```

```python
sn.catplot(x="Pclass",y="Age",kind="swarm", data=df)
sn.catplot(x="Pclass",y="Age",kind="box", data=df)
sn.catplot(x="Pclass",y="Age",kind="violin", data=df)
```

```python
sn.catplot(x="Pclass",y="Age",kind="violin",split=True, data=df)
sn.catplot(x="Pclass",y="SibSp",kind="violin",split=True, data=df)
```

```python
sn.catplot(x="Pclass",y="Age",kind="bar",data=df)
```
